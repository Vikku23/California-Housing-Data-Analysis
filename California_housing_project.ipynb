This dataset includes details about California housing data so that we can know here what the average home price will be in California.

This dataset consists of the following features are as follows:

longitude: A measure of how far west a house is; a higher value is farther west: longitude is continuous data.

latitude: A measure of how far north a house is; a higher value is farther north : latitude is continuous data

Housing median Age: Median age of a house within a block; a lower number is a newer building. It is discrete data.

Total Rooms: Total number of rooms within a block. It is discrete data.

Total Bedrooms: Total number of bedrooms within a block. It is discrete data

Population: Total number of people residing within a block. It is discrete data

House holds: Total number of households, a group of people residing within a home unit, for a block. It is discrete data.

Median Income: Median income for households within a block of houses (measured in tens of thousands of US Dollars). It comes under continuous data.

Median House Value: Median house value for households within a block (measured in US Dollars). It comes under continuous data.

Ocean Proximity: Location of the house w.r.t ocean/sea. This is categorical type of data.

6. Write a programming construct (create a user defined function) to calculate the median value of the data set wherever required.

```python
def calculate_median(data):
    sorted_data = sorted(data)  # we need to sort the data in ascending order
    n = len(sorted_data)

    if n % 2 == 1:
        median = sorted_data[n // 2]  # If the number is odd, the median is the middle value
    else:
        mid1 = sorted_data[n // 2 - 1]   # If the number  is even, the median is the average of the two middle values
        mid2 = sorted_data[n // 2]
        median = (mid1 + mid2) / 2

    return median


result1 = calculate_median(house.longitude)
print("Median of longitude:", result1)

result2 = calculate_median(house.latitude)
print("Median of latitude:", result2)

result3 = calculate_median(house.housing_median_age)
print("Median of housing_median_age :", result3)

result4 = calculate_median(house.total_rooms)
print("Median of total_rooms:", result4)

result5 = calculate_median(house.total_bedrooms)
print("Median of total_bedrooms:", result5)

result6 = calculate_median(house.population)
print("Median of population:", result6)

result7 = calculate_median(house.households)
print("Median of households:", result7)

result8 = calculate_median(house.median_income)
print("Median of median_income:", result8)

result9 = calculate_median(house.median_house_value)
print("Median of median_house_value:", result9)

```

10. Please create a new column named total_bedroom_size. If the total bedrooms is 10 or less, it should be quoted as small. If the total bedrooms is 11 or more but less than 1000, it should be medium, otherwise it should be considered large.

```python
def bedroom_size(total_bedrooms):  ## define a function bedroom size
  if total_bedrooms <= 10:
    return "small"
  elif total_bedrooms >= 11 or  total_bedrooms < 1000:
    return "medium"
  else:
    return "large"

house['total_bedroom_size'] = house['total_bedrooms'].apply(bedroom_size) ##apply the fuction to create new column

print(house)
```
